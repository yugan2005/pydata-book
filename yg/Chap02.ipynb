```python
from collections import defaultdict
```

```python
def get_counts(sequence):
    counts = defaultdict(int)
    for record in sequence:
        counts[record] += 1
    return counts
```

```python
def top_counts(count_dict, n=10):
    key_value_pairs = [(value, key) for key, value in count_dict.items()]
    key_value_pairs.sort()
    return key_value_pairs[-n:]
```

```python
years = range(1880, 2011)
pieces = []
columns=['name', 'sex', 'births']
for year in years:
    filepath = path+"yob%d.txt" %year
    frame = pd.read_csv(filepath, names=columns)
    frame['year']=year
    pieces.append(frame)
names = pd.concat(pieces, ignore_index=True)
```

```python
names[-10:]
```

```python
total_births = names.pivot_table(values='births', index='year', columns='sex', aggfunc=sum)
```

```python
total_births.tail()
```

```python
total_births.plot(title="Total births by sex and year")
```

```python
def add_prop(group):
    births = group.births.astype(float)
    group['prop'] = births / births.sum()
    return group

names = names.groupby(['year', 'sex']).apply(add_prop)
```

```python
names
```

```python
np.allclose(names.groupby(['year', 'sex']).prop.sum(), 1)
```

```python
def get_top1000(group):
    return group.sort_index(by='births', ascending=False)[:1000]
```

```python
grouped = names.groupby(['year', 'sex'])
type(grouped)
```

```python
def get_quantile_count(group, q=0.5):
    group=group.sort_index(by='prop', ascending=False)
    group_cumsum = group.prop.cumsum()
    index = group_cumsum.searchsorted(q)[0]+1
    return index
```

```python
diversity = top1000.groupby(['year', 'sex']).apply(get_quantile_count)
diversity = diversity.unstack('sex')
```
